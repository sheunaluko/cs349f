```python


class three_window_sine(BacktestTrader): 
    def __init__(self, bin_interval_ms=500):
        # Initialize Trader.
        BacktestTrader.__init__(self,  bin_interval_ms=bin_interval_ms)
        
    def calc_average_derivative(chunk) : 
        return chunk.diff().mean()

    def algorithm(self, df, **kwargs):
        
        window = kwargs['window'] #window length 
        smooth = kwargs['smooth'] #smoothing length 
        
        # Handle case when data not available
        if len(df) < window*3 + smooth :
            return None , None 

        # grab relevant data 
        df = df.iloc[-window*3-smooth:]  
        curr_price = df.iloc[-1]['ClosePrice']
        #smooth it 
        df.loc[:,'smoothed'] = df['ClosePrice'].rolling(smooth).mean() 
        
        #get chunks trying to detect something like...   \  _   / 
        chunk1  = df['smoothed'].iloc[-window*3:-window*2]
        chunk2 = df['smoothed'].iloc[-window*2:-window]
        chunk3  = df['smoothed'].iloc[-window:]

        d1, d2, d3 = [chunk.diff().mean() for chunk in [chunk1, chunk2, chunk3]]
        
        seq = [ (d2 > d1 ), (d3 > d2 ) ] 
        
        if seq == [True , True] : 
            # two increases, pass 
            return None, None 
        elif seq == [True , False]  : 
            # sine top 
            return 'Sell' , curr_price
        elif seq == [False, True] : 
            # sine bottom 
            return 'Buy' , curr_price
        else : 
            #two decreases, pass 
            return None, None 
        
class n_step_momentum(BacktestTrader): 
    def __init__(self, bin_interval_ms=500):
        # Initialize Trader.
        BacktestTrader.__init__(self,  bin_interval_ms=bin_interval_ms)

    def algorithm(self, df, **kwargs):
        
        n = kwargs['n']
        # Handle case when data not available
        if len(df) < n + 1 :
            return None, None

        curr_price = df.iloc[-1]['ClosePrice']
        prev_price = df.iloc[-n]['ClosePrice']
        
        if curr_price > prev_price : 
            return 'Buy', curr_price 
        elif curr_price < prev_price : 
            return 'Sell', curr_price
        else:
            return None, None
        
        
        
        
```

```python
algo_bank = { 
    'nsm4' : (n_step_momentum, {'n' : 4}) , 
    'nsm10' : (n_step_momentum, {'n' : 10}) , 
    'sine1' : (three_window_sine , {'window' : 40  , 'smooth' : 10 }) , 
    'sine2' : (three_window_sine , {'window' : 100  , 'smooth' : 10 }) , 
    'sine3' : (three_window_sine , {'window' : 200  , 'smooth' : 10 }) , 
}
```

```python
def backtest_algo_on_data(algo, data,bin_interval=500) : 
    
    # Initial capital and shares we will use for the backtest. 
    init_capital = 100000
    init_shares = 1000
    num_shares = 10
    
    # create the instance 
    print(algo)
    strategy,params = algo 
    trader = strategy(bin_interval_ms=bin_interval)
    
    # do the backtest 
    num_shares = 10
    init_capital
    return trader.backtest(data, num_shares, init_capital,
                 init_shares, **params)

def run_algo_bank(bank,data,bin_interval=500) : 
    results = [] 
    for _id, algo in bank.items() : 
        roi, actions = backtest_algo_on_data(algo,data,bin_interval=bin_interval)
        results.append((_id,roi,actions))
    #sort by roi
    results.sort(key=lambda x:x[1],reverse=True)
    #return 
    return results     
```

```python
results = run_algo_bank(algo_bank, cc_data)
rois = [ (x[0],x[1],i) for (i,x) in enumerate(results) ]
```

```python
windows = [5,10,20,40,100,200,300,500]
smooths = [1,5,10,25,100]

# create the algobank 
algobank = {} 
for window in windows  : 
    for smooth in smooths  : 
        algobank['sine_{}_{}'.format(window,smooth)]  = ( three_window_sine , {'window' : window  , 'smooth' : smooth } ) 
        
# backtest 
results = run_algo_bank(algobank, cc_data)
rois = [ (x[0],x[1],i) for (i,x) in enumerate(results) ]
```
